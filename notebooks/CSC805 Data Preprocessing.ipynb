```python
dataset.drop(dataset[dataset['genre'] == "set()"].index, inplace=True)

for i in dataset.index:
    genres = dataset.at[i, 'genre'].split(", ")
    if len(genres) > 1:
        dataset.at[i, 'genre'] = genres[0]

dataset['genre'].value_counts()
```
